```python
G.add_edge(1, 2)
e = (2, 3)
G.add_edge(*e)  # unpack edge tuple*
```

```python
print(G.nodes()) # returns a list
```

we add new nodes/edges and NetworkX quietly ignores any that are already present.


```python
G.add_edges_from([(1, 2), (1, 3)])
G.add_node(1)
G.add_edge(1, 2)
G.add_node("spam")        # adds node "spam"
G.add_nodes_from("spam")  # adds 4 nodes: 's', 'p', 'a', 'm'
G.add_edge(3, 'm')
```

```python
print(G.nodes()) # returns a list
```

```python
print(G.edges()) # returns list of edges
```

```python
print(G.degree()) # returns list of degree of vertices
```

```python
print(G.adj['s']) # returns adjacent nodes of 's'
```

```python
print(G.adj[1]) 
```

creating a weighted graph
a weighted graph represents a graph with weighted edges. In general, we consider the edge weights as non-negative numbers.
In the following example, E is a Python list, which contains five elements. Each of these elements is a Python tuple having three elements. The first two elements denote the two endpoints of an edge and the third element represents the weight of that edge. Finally, we need to add these weighted edges to G.

```python
G.clear() # deletes all nodes and edges
G = nx.Graph() 
E = [('A', 'B', 2), ('A', 'C', 1), ('B', 'D', 5), ('B', 'E', 3), ('C', 'E', 2)]
G.add_weighted_edges_from(E)
```

```python
pos=nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, font_weight='bold', node_color='g', node_size=700)
edge_weight = nx.get_edge_attributes(G,'weight')
print(nx.dijkstra_path(G, 'A', 'E'))
nx.draw_networkx_edge_labels(G, pos, edge_labels = edge_weight)
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.karate_club_graph()
nx.info(G)
print("Node   Degree")
for v in G:
    print(f"{v:4} {G.degree(v):7}")

nx.draw_circular(G, with_labels=True)
plt.show()
```
